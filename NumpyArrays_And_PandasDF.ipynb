```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
array = np.array([[1, 2, 3], [4, 5, 6]])
print("Original Array:")
print(array)

print("Shape of array:", array.shape)
reshaped = array.reshape((3, 2))
print("Reshaped Array (3x2):")
print(reshaped)

print("Array multiplied by 2:")
print(array * 2)

random_arr = np.random.rand(3, 3)
print("Random 3x3 Array:")
print(random_arr)

print("Second row:", array[1])
print("First column:", array[:, 0])
```

```python
data = {
    'Name' : ['Alice', 'Bob', 'Charlie', 'David'],
    'Age' : [25, 30, 35, 40],
    'Salary' : [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)
print("Head of DataFrame:")
print(df.head())

print("Describe:")
print(df.describe())

print("Info:")
df.info()

print("Select 'Age' column:")
print(df['Age'])

print("Select first two rows:")
print(df.iloc[:2])
```

```python
#line plot
plt.figure(figsize=(6,4))
plt.plot(df['Age'], df['Salary'], marker='o', linestyle='--', color='green')
plt.title("Salary vs Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
#bar plot

plt.figure(figsize=(6,4))
plt.bar(df['Name'], df['Salary'], color='skyblue')
plt.title("Salaries of Employees")
plt.xlabel("Employee")
plt.ylabel("Salary")
plt.show()
```

```python
#histogram
plt.figure(figsize=(6, 4))
plt.hist(df['Age'], bins=5, color='purple', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df['Age'], df['Salary'], color='red')
plt.title("Age vs Salary (Scatter)")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
